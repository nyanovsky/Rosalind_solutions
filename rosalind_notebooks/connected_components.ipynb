```python
def graph(txt):
    adj_list = []
    for line in txt:
        adj_list.append(line.strip())
    n = int(adj_list[0])
    adj_list = [tuple(edge.split(' ')) for edge in adj_list[1:]]
    return adj_list, n
```

```python
f = open('D:/rosalind/rosalind_tree (3).txt', 'r')
```

```python
edges, n  = graph(f)
nodes = list(range(1,n+1))
edges = [(int(x),int(y)) for x,y in edges]
```

```python
def dfs(adj, node, status):
    vecinos = []
    for x,y in adj:
        if x == node:
            vecinos.append(y)
        elif y == node:
            vecinos.append(x)
    for vecino in vecinos:
        if status[vecino] == 0:
            status[vecino] = 1
            dfs(adj, vecino, status)
```

```python
stat = {}
for node in nodes:
    stat[node] = 0
```

```python
stat
```

```python
connected = 0
for node in nodes:
    if stat[node] == 0:
        stat[node] = 1
        connected += 1
        dfs(edges,node, stat)
        
```
