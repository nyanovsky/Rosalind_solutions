```python
import pandas as pd
import numpy as np
```

```python
f = open('D:/rosalind/rosalind_cons (1).txt', 'r')
```

```python
data = pd.read_csv(f)
```

```python
raw = data.to_numpy()
```

```python
raw
```

```python
raw_ = []
s = ''
for i in range(len(raw)):
    if raw[i][0].startswith('>'):
        raw_.append(s)
        s = ''
    elif i == len(raw)-1:
        s += raw[i][0]
        raw_.append(s)
    else:
        s += raw[i][0]
```

```python
raw_
```

```python
dna =[list(x) for x in raw_]
```

```python
dna = np.array(dna)
```

```python
dna = pd.DataFrame(dna)
```

```python
bases = ['A', 'C', 'G', 'T']
```

```python
profile = pd.DataFrame(columns=dna.columns, index=bases)
```

```python
for column in dna.columns:
    profile[column] = [list(dna[column]).count(base) for base in bases]
```
