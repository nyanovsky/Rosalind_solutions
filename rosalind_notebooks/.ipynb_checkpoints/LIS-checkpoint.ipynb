```python
def LIS(arr):
    n = len(arr)
    max_arr = [1]*n
    res = []
    for i in range(n):
        for j in range(i):
            if arr[i] > arr[j] and max_arr[i] < max_arr[j]+1:
                max_arr[i] = max_arr[j]+1
    max_idx = max_arr.index(max(max_arr))
    res.insert(0,arr[max_idx])
    curr = max_arr[max_idx]
    for k in range(max_idx-1,-1,-1):
        if max_arr[k] == curr-1:
            res.insert(0,arr[k])
            curr = max_arr[k]   
    return res
```
