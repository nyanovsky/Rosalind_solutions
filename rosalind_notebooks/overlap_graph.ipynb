```python
import pandas as pd
import numpy as np
```

```python
f = open('D:/rosalind/rosalind_grph.txt')
```

```python
def to_dict(fasta):
    data = pd.read_csv(fasta, header=None)
    raw = data.to_numpy()
    d = {}
    s = ''
    i = 0
    while i < len(raw):
        if raw[i][0].startswith('>'):
                d[raw[i][0]] = ''
                j = i+1
                while j < len(raw) and raw[j][0][0] != '>':
                    s += raw[j][0]
                    j += 1
                d[raw[i][0]] = s
                s = ''
                i = j
    return d
            
```

```python
fasta = to_dict(f)
```

```python
def Ogk(k,fasta_dict):
    adj = []
    keys = list(fasta_dict.keys())
    for i in range(len(keys)):
        for j in range(len(keys)):
            if i != j:
                if fasta_dict[keys[i]][-k:] == fasta_dict[keys[j]][:k]:
                    adj.append((keys[i], keys[j]))
    return adj
        
```

```python
Ogk(3,fasta)
```

```python
for v,w in Ogk(3,fasta):
    print(v[1:],w[1:])
```
